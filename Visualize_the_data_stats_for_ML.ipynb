```python
import pandas
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pandas.read_csv('pima-indians-diabetes.csv', names=names)
peek = data.head(20)
print(peek)
```

```python
#data.drop('Unnamed: 0', inplace=True, axis=1)
data.describe()
```

```python
## Peek at the first 20 rows of data
peek = data.head(20)
print(peek)

## Shape of your data
shape = data.shape
print(shape)

## Types of data available in your dataset
types = data.dtypes
print(types)

## Statistical summary of the data set
description = data.describe()
print(description)
```

```python
## For classification ML problems let's look at the class distribution (i.e. how balanced our dataset is)
class_counts = data.groupby('class').size()
print(class_counts)
```

```python
## Let's check th Pearson's Correlation coefficient between the features
pandas.set_option('display.width', 100)
pandas.set_option('precision', 3)
correlations = data.corr(method='pearson')
print(correlations)
```

```python
## Check the skewness (distribution behavior) of our data set/ individual parameter 
#### This is more suitable for univariate problems
skew = data.skew()
print(skew)
```
