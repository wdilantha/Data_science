# Univariate PLOTS

# Histogram

```python
import matplotlib.pyplot as plt
import pandas
#url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pandas.read_csv('pima-indians-diabetes.csv', names=names)

fig = plt.figure(figsize = (10,12))
ax = fig.gca()
data.hist(ax = ax)
plt.show()
```

## Density PLOT 

```python
# Univariate Density Plots
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False,figsize=(10,15))
plt.show()
```

## BOX PLOTS

```python
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(10,15))
plt.show()
```

# Multivariate PLOTS

## Correlation PLOTS 

```python
# Correction Matrix Plot
import matplotlib.pyplot as plt
import pandas
import numpy

correlations = data.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

## Scatter plot matrix 

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data,figsize=(10,15))
plt.show()
```
